```python
π  = pi
a0 = 4
```

$
\begin{align}
\begin{bmatrix}
a_{1x}&a_{1y}&a_{1z} \\
a_{2x}&a_{2y}&a_{2z} \\
a_{3x}&a_{3y}&a_{3z}
\end{bmatrix} 
\begin{bmatrix}
b_{1x}&b_{1y}&b_{1z} \\
b_{2x}&b_{2y}&b_{2z} \\
b_{3x}&b_{3y}&b_{3z}
\end{bmatrix} =
\begin{bmatrix}
2\pi& 0 & 0\\
0 & 2\pi & 0\\
0 & 0 &2\pi
\end{bmatrix}
\end{align}
$

```python
# Vectores del espacio real
a1 = array([1,0,0])*a0
a2 = array([0,1,0])*a0
a3 = array([0,0,1])*a0

A  = array( [a1,a2,a3] )
B  = 2*π*linalg.inv(A)

# Vectores del espacio recíproco
b1 = B[:,0]
b2 = B[:,1]
b3 = B[:,2]
```

## Puntos de la red
$
\begin{align}
P = u_1\vec{a}_1 + u_2\vec{a}_2 + u_3\vec{a}_3
\end{align}
$

donde $u_i=-n,-n+1,\ldots,n-1,n$

```python
n = 2
P = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        for u3 in arange(-n,n+1):
            P.append( u1*a1+u2*a2+u3*a3 )
P = array(P)
```

## Puntos de la red recíproca
$
\begin{align}
G = v_1\vec{b}_1 + v_2\vec{b}_2 + v_3\vec{b}_3
\end{align}
$

donde $v_i=-n,-n+1,\ldots,n-1,n$

```python
G = []
for v1 in arange(-n,n+1):
    for v2 in arange(-n,n+1):
        for v3 in arange(-n,n+1):
            G.append( v1*b1+v2*b2+v3*b3 )
G = array(G)
```

```python
fig = plt.figure(figsize=(15,5))
ax_real = fig.add_subplot(1, 3, 1, projection='3d')

ax_real.scatter(P.T[0],P.T[1],P.T[2],c='k')


ax_recip = fig.add_subplot(1, 3, 2, projection='3d')
ax_recip.scatter(G.T[0],G.T[1],G.T[2],c='r')

ax_both = fig.add_subplot(1, 3, 3, projection='3d')
ax_both.scatter(P.T[0],P.T[1],P.T[2],c='k')
ax_both.scatter(G.T[0],G.T[1],G.T[2],c='r')
```

# SC
---
### El azufre cristaliza en una red cúbica simple, por lo que tiene un átomo por cada punto de la red. Pongamos cada átomo justo en cada punto de la red.

$
\begin{align}
S_{\vec{G}} = \sum_j f_j e^{-i\vec{G}\cdot\vec{r_j}} = f e^{-i\vec{G}\cdot\vec{P}}
\end{align}
$

```python
f  = 4
θ  = array( [ -1J*dot(G[n],P[n]) for n in range(len(G)) ] )
SG = f*exp(θ)
I  = real(conj(SG)*SG)
```

```python
I
```

```python
fig = plt.figure(figsize=(15,5))
ax_real = fig.add_subplot(1, 2, 1, projection='3d')

ax_real.scatter(P.T[0],P.T[1],P.T[2],c='k')

ax_recip = fig.add_subplot(1, 2, 2, projection='3d')
ax_recip.scatter(G.T[0],G.T[1],G.T[2],c=I,cmap='Reds_r' )
```

# BCC

```python
# Vectores del espacio real
a1 = array([1,0,0])*a0
a2 = array([0,1,0])*a0
a3 = array([0,0,1])*a0

A  = array( [a1,a2,a3] )
B  = 2*π*linalg.inv(A)

# Vectores del espacio recíproco
b1 = B[:,0]
b2 = B[:,1]
b3 = B[:,2]
```

```python
n = 2
r1 = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        for u3 in arange(-n,n+1):
            r1.append( u1*a1+u2*a2+u3*a3 + array([0,0,0]))
r1 = array(r1)

r2 = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        for u3 in arange(-n,n+1):
            r2.append( u1*a1+u2*a2+u3*a3 + array([1/2,1/2,1/2])*a0 )
r2 = array(r2)
```

```python
G = []
for v1 in arange(-n,n+1):
    for v2 in arange(-n,n+1):
        for v3 in arange(-n,n+1):
            G.append( v1*b1+v2*b2+v3*b3 )
G = array(G)
```

```python
fig = plt.figure(figsize=(15,5))
ax_real = fig.add_subplot(1, 3, 1, projection='3d')

ax_real.scatter(r1.T[0],r1.T[1],r1.T[2],c='k')
ax_real.scatter(r2.T[0],r2.T[1],r2.T[2],c='g')


ax_recip = fig.add_subplot(1, 3, 2, projection='3d')
ax_recip.scatter(G.T[0],G.T[1],G.T[2],c='r')

ax_both = fig.add_subplot(1, 3, 3, projection='3d')
ax_both.scatter(r1.T[0],r1.T[1],r1.T[2],c='k')
ax_both.scatter(r2.T[0],r2.T[1],r2.T[2],c='g')
ax_both.scatter(G.T[0],G.T[1],G.T[2],c='r')
```

$
\begin{align}
S_{\vec{G}} = \sum_j f_j e^{-i\vec{G}\cdot\vec{r_j}} = f e^{-i\vec{G}\cdot\vec{r_1}}+f e^{-i\vec{G}\cdot\vec{r_2}}
\end{align}
$

```python
f  = 4
θ1  = array( [ -1J*dot(G[n],r1[n]) for n in range(len(G)) ] )
θ2  = array( [ -1J*dot(G[n],r2[n]) for n in range(len(G)) ] )
SG = f*exp(θ1)+f*exp(θ2)
I  = real(conj(SG)*SG)
```

```python
I
```

```python
fig = plt.figure(figsize=(15,5))
ax_real = fig.add_subplot(1, 2, 1, projection='3d')

ax_real.scatter(r1.T[0],r1.T[1],r1.T[2],c='k')
ax_real.scatter(r2.T[0],r2.T[1],r2.T[2],c='g')

ax_recip = fig.add_subplot(1, 2, 2, projection='3d')
ax_recip.scatter(G.T[0],G.T[1],G.T[2],c=I,cmap='Reds_r')
```

```python
G = []
for v1 in arange(-n,n+1):
    for v2 in arange(-n,n+1):
        for v3 in arange(-n,n+1):
            G.append( v1*b1+v2*b2+v3*b3 )
G = array(G)
```

$
\begin{align}
S_{\vec{G}} = \sum_j f_j e^{-i\vec{G}\cdot\vec{r_j}} = f e^{-i\vec{G}\cdot\vec{r_1}}+f e^{-i\vec{G}\cdot\vec{r_2}}
+f e^{-i\vec{G}\cdot\vec{r_3}}+f e^{-i\vec{G}\cdot\vec{r_4}}
\end{align}
$

```python
f  = 4; 
θ1  = array( [ -1J*dot(G[n],r1[n]) for n in range(len(G)) ] )
θ2  = array( [ -1J*dot(G[n],r2[n]) for n in range(len(G)) ] )
θ3  = array( [ -1J*dot(G[n],r3[n]) for n in range(len(G)) ] )
θ4  = array( [ -1J*dot(G[n],r4[n]) for n in range(len(G)) ] )
SG = f*exp(θ1)+f*exp(θ2)+f*exp(θ3)+f*exp(θ4)
I  = real(conj(SG)*SG)
```

```python
n = 2
r1_SC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r1_SC.append( u1*a1+u2*a2+u3*a3 + array([0,0,0]))
r1_SC = array(r1_SC)
```

```python
n = 2
r1_BCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r1_BCC.append( u1*a1+u2*a2+u3*a3 + array([0,0,0]))
r1_BCC = array(r1_BCC)

r2_BCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r2_BCC.append( u1*a1+u2*a2+u3*a3 + array([1/2,1/2,1/2])*a0 )
r2_BCC = array(r2_BCC)
```

```python
n = 2
r1_FCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r1_FCC.append( u1*a1+u2*a2+u3*a3 + array([  0,  0,  0])*a0 )
r1_FCC = array(r1_FCC)

r2_FCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r2_FCC.append( u1*a1+u2*a2+u3*a3 + array([  0,1/2,1/2])*a0 )
r2_FCC = array(r2_FCC)

r3_FCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r3_FCC.append( u1*a1+u2*a2+u3*a3 + array([1/2,  0,1/2])*a0 )
r3_FCC = array(r3_FCC)

r4_FCC = []
for u1 in arange(-n,n+1):
    for u2 in arange(-n,n+1):
        # for u3 in arange(-n,n+1):
            u3 = 1
            r4_FCC.append( u1*a1+u2*a2+u3*a3 + array([1/2,1/2,  0])*a0 )
r4_FCC = array(r4_FCC)
```

```python
G = []
for v1 in arange(-n,n+1):
    for v2 in arange(-n,n+1):
        # for v3 in arange(-n,n+1):
            v3 = 1
            G.append( v1*b1+v2*b2+v3*b3 )
G = array(G)
```

```python
f  = 4
θ1  = array( [ -1J*dot(G[n],r1_SC[n]) for n in range(len(G)) ] )
SG_SC = f*exp(θ1)
I_SC  = real(conj(SG_SC)*SG_SC)
```
